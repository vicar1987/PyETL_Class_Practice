```python
import jieba
import os
import glob

jieba.load_userdict('./mydict.txt')
stopwordList = list()
with open('./stopword.txt', 'r', encoding='utf-8') as f:
    stopwordList = [w.replace('\n', '') for w in f.readlines()]

foldername = 'pttGossiping'
os.listdir(foldername)
sb = ''
for fileName in os.listdir(foldername):
    path = foldername + '/' + fileName
#     print(path)
    with open(path, 'r', encoding='utf-8') as f:
        sb += f.read().split('---split---')[0]
        
# for fileName in glob.glob(foldername + ):
#     path = foldername + '/' + fileName
# #     print(path)
#     with open(path, 'r', encoding='utf-8') as f:
#         sb += f.read().split('---split---')[0]
        
articleCut = jieba.cut(sb)
wordList = list()
for w in articleCut:
    wordList.append(w)


wordCountDict = dict()
for w in wordList:
    if (w.__len__() < 2) or (w in stopwordList):
        continue
#     if w in stopwordList:
#         continue
    if w in wordCountDict:
        wordCountDict[w] += 1
    else:
        wordCountDict[w] = 1

wordCountList = list()
for k, v in wordCountDict.items():
    wordCountList.append((k, v))
    
wordCountList.sort(key=lambda x: x[1], reverse=True)
wordCountList
```
