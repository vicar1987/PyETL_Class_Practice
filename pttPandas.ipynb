```python
import pandas as pd
import glob

folderName = 'pttGossiping'
# glob.glob(folderName + './*txt')

data = list()
for path in glob.glob(folderName + './*txt'):
    with open(path, 'r', encoding='utf-8') as f:
        tmpList = f.read().split('---split---')[1].split('\n')[1:-1]
        data.append(tmpList)
        
# data

columns = [r.split(': ')[0] for r in data[0]]
# columns

df = pd.DataFrame(data=data, columns=columns)
df
```

```python
def columnFilter(x):
    output = x.split(': ')[1]
    return output

df['推'].apply(columnFilter)
```

```python
df2 = df.copy()
```

```python
for c in df:
    df[c] = df[c].apply(columnFilter)
```

```python
df
```

```python
df2['推'].apply(lambda x: x.split(': ')[1]) # 簡單的轉換邏輯直接用lambda解決
```

```python
for c in df2:
    df2[c] = df2[c].apply(lambda x: x.split(': ')[1])
```
